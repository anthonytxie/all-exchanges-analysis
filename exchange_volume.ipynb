```python
import ccxt
import numpy as np
import pandas as pd
import time
import json
import threading
```

```python

exchange_names = ['kucoin', 'kraken', 'coinbase', 'coinbasepro', 'bittrex']
exchanges = [ccxt.kucoin(), ccxt.kraken(), ccxt.coinbase(), ccxt.coinbasepro(), ccxt.bittrex()]






for i, name in enumerate(exchange_names):
    exchanges[i].load_markets()
    exchange = exchanges[i]
    for symbol in exchange.markets:
        time.sleep (exchange.rateLimit / 1000) # time.sleep wants seconds
        ohlcv = exchange.fetch_ohlcv (symbol, '1w', since=1567192975437)
        for datum in ohlcv:
            fat_dict['date'].append(datum[0])
            fat_dict['exchange'].append(name)
            fat_dict['symbol'].append(symbol)
            fat_dict['volume'].append(datum[5])




```
