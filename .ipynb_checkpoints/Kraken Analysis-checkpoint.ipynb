```python
import pandas as pd 
import numpy as np
import json
import os
from datetime import datetime
```

```python
file_array = []
dates = []


for file in os.listdir("/Users/Anthony/desktop/exchange-orderbooks"):
    if file.endswith("kraken_orderbook.json"):
        file_array.append(os.path.join("/Users/Anthony/desktop/exchange-orderbooks", file))


for file in file_array:
    date = file.split('/')[5].split('_')[0]
    dates.append(date)
    


dates
```

```python
slippage_pairs = []


for file in file_array:
    try: 
        with open(file, encoding='utf-8') as data_file:
            dataset = json.loads(data_file.read())
        for exchange_pair in dataset:
            slippage_dict = {}
            date = datetime.fromtimestamp(int(file.split('/')[5].split('_')[0])/1000).strftime('%Y-%m-%d %H:%M:%S')
            slippage_dict['date'] = date
            slippage_dict['exchange_name'] = 'kraken'
            slippage_dict['pair'] = exchange_pair 
            slippage_dict['bids'] = dataset[exchange_pair]['bids']
            slippage_dict['asks'] = dataset[exchange_pair]['asks']
            slippage_dict['base_currency'] = exchange_pair.split('/')[1]
            if len(slippage_dict['bids']) > 15 and len(slippage_dict['asks']) > 15:
                slippage_pairs.append(slippage_dict)
            else:
                print('not long enough')
                continue
    except: 
        continue



slippage_pairs[0]
```

```python

def calc_slippage_asks(asks, base_currency):
    asking_price = slippage_pairs[0]['asks'][0][0]
    index = 0
    if base_currency == 'BTC':
        buying_amount = 0.47
    elif base_currency =='ETH':
        buying_amount = 28.71
    elif base_currency == 'EUR':
        buying_amount = 4532.50
    elif base_currency == 'KRW':
        buying_amount = 5985900
    elif base_currency == 'CAD':
        buying_amount = 6616
    elif base_currency == 'JPY':
        buying_amount = 535067
    elif base_currency == 'GDP':
        buying_amount = 4057
    else: 
        buying_amount = 5000
    while buying_amount > 0:
        try:
            amount = asks[index][0] * [asks][index][1]
            if buying_amount - amount < 0:
                break
            else:
                index +=1
        except:
            return 100

    return (np.absolute((asking_price - asks[index][0]) / asking_price)) * 100


def calc_slippage_bids(bids, base_currency):
    bid_price = bids[0][0]
    index = 0
    if base_currency == 'BTC':
        selling_amount = 0.47
    elif base_currency =='ETH':
        selling_amount = 28.71
    elif base_currency == 'EUR':
        selling_amount = 4532.50
    elif base_currency == 'KRW':
        selling_amount = 5985900
    elif base_currency == 'CAD':
        selling_amount = 6616
    elif base_currency == 'JPY':
        selling_amount = 535067
    elif base_currency == 'GDP':
        selling_amount = 4057
    else: 
        selling_amount = 5000
        
    while selling_amount > 0:
        try:
            print(bids[index][0], bids[index][1])
            amount = bids[index][0] * bids[index][1]
            if selling_amount - amount < 0:
                break
            else:
                index +=1
        except:
            return 100
    return ((bids[index][0] - bid_price) / bid_price) * 100



        

```

```python
slippage_array = []

for index, pair in enumerate(slippage_pairs):
    print(index)
    slippage_dict = {}
    slippage_dict['exchange_name'] = pair['exchange_name']
    slippage_dict['date'] = pair['date']
    slippage_dict['bid_slippage'] = calc_slippage_asks(pair['bids'], pair['base_currency'])
    slippage_dict['ask_slippage'] = calc_slippage_bids(pair['asks'], pair['base_currency'])
    slippage_dict['pair'] = pair['pair']
    slippage_array.append(slippage_dict)
```

```python
slippage_df = pd.DataFrame(slippage_array)
```

```python
unique_base_currencies = set()


for key in pd.DataFrame(slippage_df['pair'].value_counts()).index:
    unique_base_currencies.add(key.split('/')[1])
    
print(unique_base_currencies)
```
